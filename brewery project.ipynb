## Imputing Missing Values With KNN Imputer
- It seems that reporting ABV and IBU can come down in some instances to the decision of the brewery given different government regulations etc.
- There is likely an argument for filling NA with 0 since it is not reported, but since certain types of drinks would likley have characteristics in terms of style, then we will impute these values given KNN.

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors = 5)
cols_to_impute = ['ABV','IBU']
beer_df[cols_to_impute] = imputer.fit_transform(beer_df[cols_to_impute])
```

```python
beer_df.dropna(inplace=True, axis = 0)
```
